4. Let's go back a couple of steps. What is a pipeline and what do we use it for?

A pipeline is a program which runs multiple individual tools on a given input to process input data in a pre-determined order with pre-chosen parameters. 

Pipelines are used to process and analyze data in a single step, where the same version and parameters of each individual tool are chosen, which significantly improved reproducability, and simplifies the data analysis, especially on larger data sets.

5. Why do you think nf-core adheres to strict guidelines?

Ensure that all pipelines in the project adhere to common standards, such as open-source, functionality, and ease of use, which boosts reproducability.

6. What are the main features of nf-core pipelines?

portability, interopability, reproducability
